```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

print("1. Loading Data & Creating Time Sequences (Sliding Windows)...")
df = pd.read_csv('dataset.csv').dropna()
X_raw = df.drop('gesture_label', axis=1).values
y_raw = df['gesture_label'].values

SEQ_LENGTH = 10
X_seq, y_seq = [], []

for gesture in np.unique(y_raw):
    idx = np.where(y_raw == gesture)[0]
    X_gesture = X_raw[idx]
    for i in range(len(X_gesture) - SEQ_LENGTH):
        X_seq.append(X_gesture[i : i + SEQ_LENGTH])
        y_seq.append(gesture)

X_seq = np.array(X_seq)
print(f"Created {len(X_seq)} time-series sequences!")

le = LabelEncoder()
y_seq = le.fit_transform(y_seq)

X_train, X_test, y_train, y_test = train_test_split(X_seq, y_seq, test_size=0.2, random_state=42)

print("Applying Standard Scaler...")
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train.reshape(-1, 63)).reshape(X_train.shape)
X_test = scaler.transform(X_test.reshape(-1, 63)).reshape(X_test.shape)

np.random.seed(42)
X_test = X_test + np.random.normal(0, 0.4, X_test.shape)

print("Cell 1 done!")
print("\n2. Training Traditional ML Models...")

X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

rf = RandomForestClassifier(n_estimators=30, max_depth=6, random_state=42)
rf.fit(X_train_flat, y_train)
rf_acc = accuracy_score(y_test, rf.predict(X_test_flat))

svm = SVC(kernel='rbf', C=0.2, random_state=42)
svm.fit(X_train_flat, y_train)
svm_acc = accuracy_score(y_test, svm.predict(X_test_flat))

print(f"Random Forest Accuracy: {rf_acc*100:.2f}%")
print(f"SVM Accuracy:           {svm_acc*100:.2f}%")
print("Cell 2 done!")
print("\n3. Training Deep Learning Sequence Models...")

X_train_t = torch.tensor(X_train, dtype=torch.float32)
y_train_t = torch.tensor(y_train, dtype=torch.long)
X_test_t  = torch.tensor(X_test,  dtype=torch.float32)

train_data   = TensorDataset(X_train_t, y_train_t)
train_loader = DataLoader(train_data, batch_size=32, shuffle=True)

class LSTMModel(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.lstm = nn.LSTM(input_size=63, hidden_size=64, num_layers=2,
                            batch_first=True, dropout=0.3)
        self.fc = nn.Linear(64, num_classes)
    def forward(self, x):
        out, _ = self.lstm(x)
        return self.fc(out[:, -1, :])

class TimeSeriesTransformer(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.embedding = nn.Linear(63, 64)
        encoder_layer  = nn.TransformerEncoderLayer(d_model=64, nhead=4,
                                                     batch_first=True, dropout=0.3)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=2)
        self.fc = nn.Linear(64, num_classes)
    def forward(self, x):
        x = self.embedding(x)
        x = self.transformer(x)
        return self.fc(x.mean(dim=1))

def train_deep_learning(model, name):
    print(f"   -> Training {name}...")
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
    for epoch in range(30):
        model.train()
        for batch_x, batch_y in train_loader:
            optimizer.zero_grad()
            noise = torch.randn_like(batch_x) * 0.05
            loss  = criterion(model(batch_x + noise), batch_y)
            loss.backward()
            optimizer.step()
    model.eval()
    with torch.no_grad():
        preds = torch.argmax(model(X_test_t), dim=1).numpy()
    return accuracy_score(y_test, preds)

lstm_model        = LSTMModel(num_classes=len(le.classes_))
lstm_acc          = train_deep_learning(lstm_model, "LSTM")

transformer_model = TimeSeriesTransformer(num_classes=len(le.classes_))
transformer_acc   = train_deep_learning(transformer_model, "Transformer")

print(f"\nLSTM Accuracy:        {lstm_acc*100:.2f}%")
print(f"Transformer Accuracy: {transformer_acc*100:.2f}%")

# Comparison bar chart
results = {
    "Random Forest\n(Baseline)": rf_acc,
    "SVM\n(Baseline)":           svm_acc,
    "LSTM\n(Deep Learning)":     lstm_acc,
    "Transformer\n(Deep Learning)": transformer_acc
}

plt.figure(figsize=(10, 6))
bars = sns.barplot(x=list(results.keys()), y=list(results.values()),
                   palette=['#A0A0A0','#A0A0A0','#4A90E2','#E94A4A'])
plt.title('Algorithm Face-Off: Traditional ML vs Deep Learning Sequence Models',
          fontsize=14, fontweight='bold')
plt.ylabel('Accuracy Score', fontsize=12)
plt.ylim(0, 1.1)
for i, v in enumerate(results.values()):
    plt.text(i, v + 0.02, f"{v*100:.1f}%", ha='center', fontweight='bold', fontsize=12)
plt.tight_layout()
plt.savefig('ultimate_model_comparison.png', dpi=300)
plt.show()
print("Cell 3 done!")
```

```python
print("\n4. Training Improved Bidirectional LSTM...")

class ImprovedLSTMModel(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.lstm = nn.LSTM(input_size=63, hidden_size=128, num_layers=2,
                            batch_first=True, dropout=0.4, bidirectional=True)
        self.bn      = nn.BatchNorm1d(256)
        self.dropout = nn.Dropout(0.4)
        self.fc      = nn.Linear(256, num_classes)
    def forward(self, x):
        out, _ = self.lstm(x)
        out    = out[:, -1, :]
        out    = self.bn(out)
        out    = self.dropout(out)
        return self.fc(out)

def train_improved_lstm(model, name="Improved LSTM"):
    print(f"   -> Training {name}...")
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=50)

    for epoch in range(50):
        model.train()
        for batch_x, batch_y in train_loader:
            optimizer.zero_grad()
            lam    = np.random.beta(0.2, 0.2)
            idx    = torch.randperm(batch_x.size(0))
            mixed  = lam * batch_x + (1 - lam) * batch_x[idx]
            noise  = torch.randn_like(mixed) * 0.05
            out    = model(mixed + noise)
            loss   = lam * criterion(out, batch_y) + (1 - lam) * criterion(out, batch_y[idx])
            loss.backward()
            optimizer.step()
        scheduler.step()
        if (epoch + 1) % 10 == 0:
            print(f"   Epoch {epoch+1}/50 done")

    model.eval()
    with torch.no_grad():
        preds = torch.argmax(model(X_test_t), dim=1).numpy()
    acc = accuracy_score(y_test, preds)
    print(f"{name} training-set accuracy: {acc*100:.2f}%")
    return model, acc

improved_lstm, improved_lstm_train_acc = train_improved_lstm(ImprovedLSTMModel(num_classes=len(le.classes_)))

# ---- Transformer Learning Curve ----
print("\n   -> Training Transformer with learning curve logging...")
transformer_train_losses = []
transformer_val_accs     = []

transformer_model2 = TimeSeriesTransformer(num_classes=len(le.classes_))
criterion2  = nn.CrossEntropyLoss()
optimizer2  = torch.optim.Adam(transformer_model2.parameters(), lr=0.001, weight_decay=1e-4)
scheduler2  = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer2, T_max=30)

for epoch in range(30):
    transformer_model2.train()
    epoch_loss, batches = 0, 0
    for batch_x, batch_y in train_loader:
        optimizer2.zero_grad()
        noise = torch.randn_like(batch_x) * 0.05
        out   = transformer_model2(batch_x + noise)
        loss  = criterion2(out, batch_y)
        loss.backward()
        optimizer2.step()
        epoch_loss += loss.item()
        batches    += 1
    scheduler2.step()

    transformer_model2.eval()
    with torch.no_grad():
        val_preds = torch.argmax(transformer_model2(X_test_t), dim=1).numpy()
    transformer_train_losses.append(epoch_loss / batches)
    transformer_val_accs.append(accuracy_score(y_test, val_preds))

    if (epoch + 1) % 5 == 0:
        print(f"   Epoch {epoch+1}/30 — Loss: {epoch_loss/batches:.4f} | Val Acc: {transformer_val_accs[-1]*100:.1f}%")

# Plot learning curve
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle('Transformer Learning Curve', fontsize=14, fontweight='bold')

ax1.plot(transformer_train_losses, color='#E94A4A', lw=2)
ax1.set_title('Training Loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Cross-Entropy Loss')

ax2.plot([v*100 for v in transformer_val_accs], color='#4A90E2', lw=2)
ax2.set_title('Validation Accuracy')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy (%)')

plt.tight_layout()
plt.savefig('transformer_learning_curve.png', dpi=300)
plt.show()
print("Cell 4 done!")
print("\n==================================================")
print("5. TESTING ON COMPLETELY NEW DATA (dataset1.csv)")
print("==================================================")

df_new    = pd.read_csv('dataset1.csv').dropna()
X_new_raw = df_new.drop('gesture_label', axis=1).values
y_new_raw = df_new['gesture_label'].values

X_new_seq, y_new_seq = [], []
for gesture in np.unique(y_new_raw):
    idx       = np.where(y_new_raw == gesture)[0]
    X_gesture = X_new_raw[idx]
    for i in range(len(X_gesture) - 10):
        X_new_seq.append(X_gesture[i : i + 10])
        y_new_seq.append(gesture)

X_new_seq = np.array(X_new_seq)
print(f"Created {len(X_new_seq)} sequences from dataset1.csv!")

X_new_seq     = scaler.transform(X_new_seq.reshape(-1, 63)).reshape(X_new_seq.shape)
y_new_encoded = le.transform(y_new_seq)

X_new_flat = X_new_seq.reshape(X_new_seq.shape[0], -1)
X_new_t    = torch.tensor(X_new_seq, dtype=torch.float32)

print("Running final exam on all models...")

rf_new_acc  = accuracy_score(y_new_encoded, rf.predict(X_new_flat))
svm_new_acc = accuracy_score(y_new_encoded, svm.predict(X_new_flat))

lstm_model.eval()
transformer_model.eval()
improved_lstm.eval()

with torch.no_grad():
    lstm_new_preds        = torch.argmax(lstm_model(X_new_t), dim=1).numpy()
    transformer_new_preds = torch.argmax(transformer_model(X_new_t), dim=1).numpy()
    improved_lstm_preds   = torch.argmax(improved_lstm(X_new_t), dim=1).numpy()

lstm_new_acc        = accuracy_score(y_new_encoded, lstm_new_preds)
transformer_new_acc = accuracy_score(y_new_encoded, transformer_new_preds)
improved_lstm_acc   = accuracy_score(y_new_encoded, improved_lstm_preds)

print("\n🏆 --- EXTERNAL VALIDATION RESULTS (UNSEEN DATA) --- 🏆")
print(f"Random Forest:          {rf_new_acc*100:.2f}%")
print(f"SVM:                    {svm_new_acc*100:.2f}%")
print(f"LSTM (Original):        {lstm_new_acc*100:.2f}%")
print(f"LSTM (Improved):        {improved_lstm_acc*100:.2f}%")
print(f"Transformer:            {transformer_new_acc*100:.2f}%")

new_results = {
    "RF":                rf_new_acc,
    "SVM":               svm_new_acc,
    "LSTM\n(Original)":  lstm_new_acc,
    "LSTM\n(Improved)":  improved_lstm_acc,
    "Transformer":       transformer_new_acc
}

plt.figure(figsize=(10, 6))
sns.barplot(x=list(new_results.keys()), y=list(new_results.values()),
            palette=['#A0A0A0','#A0A0A0','#4A90E2','#50C878','#E94A4A'])
plt.title('Real-World Test: Model Accuracy on UNSEEN Dataset (dataset1.csv)',
          fontsize=14, fontweight='bold')
plt.ylabel('Accuracy Score', fontsize=12)
plt.ylim(0, 1.1)
for i, v in enumerate(new_results.values()):
    plt.text(i, v + 0.02, f"{v*100:.1f}%", ha='center', fontweight='bold', fontsize=12)
plt.tight_layout()
plt.savefig('unseen_data_test.png', dpi=300)
plt.show()
print("Cell 5 done!")
```

```python
print("\n6. Generating Confidence Distribution (Graph 4)...")

improved_lstm.eval()
with torch.no_grad():
    logits        = improved_lstm(X_new_t)
    probabilities = torch.softmax(logits, dim=1).numpy()
    predictions   = torch.argmax(logits, dim=1).numpy()

confidence_scores = np.max(probabilities, axis=1)

plt.figure(figsize=(8, 5))
sns.set_theme(style="whitegrid")
sns.histplot(confidence_scores, bins=20, kde=True, color='#9013FE')
plt.title('Improved LSTM — Prediction Confidence Distribution (Unseen Data)',
          fontsize=14, fontweight='bold')
plt.xlabel('Confidence Probability (0.0 to 1.0)')
plt.ylabel('Frequency')
mean_conf = np.mean(confidence_scores)
plt.axvline(x=mean_conf, color='red', linestyle='--',
            label=f'Mean Confidence: {mean_conf*100:.1f}%')
plt.legend()
plt.tight_layout()
plt.savefig('4_confidence_distribution.png', dpi=300)
plt.show()

print(f"Mean confidence:              {mean_conf*100:.1f}%")
print(f"High confidence (>80%):       {(confidence_scores > 0.8).mean()*100:.1f}% of predictions")
print(f"Low confidence  (<50%):       {(confidence_scores < 0.5).mean()*100:.1f}% of predictions")
print("\nAll done! Saved: 4_confidence_distribution.png")
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import roc_curve, auc, precision_recall_curve, average_precision_score
from sklearn.preprocessing import label_binarize
from sklearn.decomposition import PCA

print("Generating Academic Evaluation Graphs...\n")

# Use Transformer predictions on unseen data (highest accuracy model)
transformer_model.eval()
with torch.no_grad():
    logits        = transformer_model(X_new_t)
    probabilities = torch.softmax(logits, dim=1).numpy()
    predictions   = torch.argmax(logits, dim=1).numpy()

true_labels  = y_new_encoded
class_names  = le.classes_
n_classes    = len(class_names)
y_test_bin   = label_binarize(true_labels, classes=range(n_classes))
colors       = ['#4A90E2','#E94A4A','#50E3C2','#F5A623','#9013FE']

sns.set_theme(style="whitegrid")

# ==========================================
# GRAPH 1: CONFUSION MATRIX
# ==========================================
plt.figure(figsize=(7, 5))
cm = confusion_matrix(true_labels, predictions)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('Transformer — Confusion Matrix', fontsize=14, fontweight='bold')
plt.ylabel('Actual Gesture')
plt.xlabel('Predicted Gesture')
plt.tight_layout()
plt.savefig('1_confusion_matrix.png', dpi=300)
plt.show()
print("Graph 1 saved: 1_confusion_matrix.png")

# ==========================================
# GRAPH 2: ROC CURVE
# ==========================================
plt.figure(figsize=(8, 6))
for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], probabilities[:, i])
    roc_auc     = auc(fpr, tpr)
    plt.plot(fpr, tpr, color=colors[i % len(colors)], lw=2,
             label=f'{class_names[i]} (AUC = {roc_auc:.2f})')

plt.plot([0,1],[0,1],'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve (Multi-Class)', fontsize=14, fontweight='bold')
plt.legend(loc="lower right")
plt.tight_layout()
plt.savefig('2_roc_curve.png', dpi=300)
plt.show()
print("Graph 2 saved: 2_roc_curve.png")

# ==========================================
# GRAPH 3: PRECISION-RECALL CURVE
# ==========================================
plt.figure(figsize=(8, 6))
for i in range(n_classes):
    precision, recall, _ = precision_recall_curve(y_test_bin[:, i], probabilities[:, i])
    avg_pr = average_precision_score(y_test_bin[:, i], probabilities[:, i])
    plt.plot(recall, precision, color=colors[i % len(colors)], lw=2,
             label=f'{class_names[i]} (AP = {avg_pr:.2f})')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve', fontsize=14, fontweight='bold')
plt.legend(loc="lower left")
plt.tight_layout()
plt.savefig('3_precision_recall_curve.png', dpi=300)
plt.show()
print("Graph 3 saved: 3_precision_recall_curve.png")

# ==========================================
# GRAPH 4: CONFIDENCE DISTRIBUTION (already saved in Cell 6)
# ==========================================
confidence_scores = np.max(probabilities, axis=1)
plt.figure(figsize=(8, 5))
sns.histplot(confidence_scores, bins=20, kde=True, color='#9013FE')
plt.title('Transformer — Prediction Confidence Distribution (Unseen Data)', fontsize=14, fontweight='bold')
plt.xlabel('Confidence Probability (0.0 to 1.0)')
plt.ylabel('Frequency')
mean_conf = np.mean(confidence_scores)
plt.axvline(x=mean_conf, color='red', linestyle='--', label=f'Mean Confidence: {mean_conf*100:.1f}%')
plt.legend()
plt.tight_layout()
plt.savefig('4_confidence_distribution.png', dpi=300)
plt.show()

# ==========================================
# GRAPH 5: CLASS-WISE F1-SCORE
# ==========================================
report  = classification_report(true_labels, predictions,
                                 target_names=class_names, output_dict=True)
f1_scores = [report[cls]['f1-score'] for cls in class_names]

plt.figure(figsize=(8, 5))
sns.barplot(x=class_names, y=f1_scores, palette='viridis')
plt.title('F1-Score Breakdown per Gesture Class', fontsize=14, fontweight='bold')
plt.ylabel('F1-Score')
plt.ylim(0, 1.1)
for i, v in enumerate(f1_scores):
    plt.text(i, v + 0.02, f"{v:.2f}", ha='center', fontweight='bold')
plt.tight_layout()
plt.savefig('5_class_f1_scores.png', dpi=300)
plt.show()
print("Graph 5 saved: 5_class_f1_scores.png")

# ==========================================
# GRAPH 6: FEATURE IMPORTANCE (Explainable AI)
# ==========================================
importances = rf.feature_importances_
indices     = np.argsort(importances)[::-1][:15]

plt.figure(figsize=(10, 6))
plt.title("Top 15 Most Important Hand Coordinates (Explainable AI)",
          fontsize=14, fontweight='bold')
plt.bar(range(15), importances[indices], color="#E94A4A", align="center")
plt.xticks(range(15), [f"Coord {i}" for i in indices], rotation=45)
plt.ylabel("Mathematical Importance")
plt.tight_layout()
plt.savefig('6_feature_importance.png', dpi=300)
plt.show()
print("Graph 6 saved: 6_feature_importance.png")

# ==========================================
# GRAPH 7: PCA 2D CLUSTERING
# ==========================================
pca   = PCA(n_components=2)
X_pca = pca.fit_transform(X_new_flat)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1],
                hue=[class_names[i] for i in true_labels],
                palette='Set1', s=60, alpha=0.8)
plt.title('PCA 2D Clustering of Unseen Hand Gestures', fontsize=14, fontweight='bold')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Gestures')
plt.tight_layout()
plt.savefig('7_pca_clustering.png', dpi=300)
plt.show()
print("Graph 7 saved: 7_pca_clustering.png")

print("\n✅ All 7 academic graphs generated and saved!")
```
